## Generating simpler dataset for PHOC model

- To generate a simpler dataset for the PHOC model using fonts downloaded from the internet

### Directory handling

```python
annotations_file = "/home/xnmaster/DlProject/deep-learning-project-2024-ai_nndl_group_01_/Synthetic/lexicon.txt"
img_dir_train = "/home/xnmaster/DlProject/deep-learning-project-2024-ai_nndl_group_01_/Generate Dataset/Generated_Train"
img_dir_test = "/home/xnmaster/DlProject/deep-learning-project-2024-ai_nndl_group_01_/Generate Dataset/Generated_Test"
font_folder_path = '/home/xnmaster/DlProject/deep-learning-project-2024-ai_nndl_group_01_/Generate Dataset/Fonts'
```

### Generating data set

```python
fonts = ["Arial.ttf", "calibri.ttf", "COMICSANS.TTF", "helvetica.ttf"]


with open(annotations_file, "r") as file:
    list_of_words = file.readlines()

list_of_words = [l[:-1] for l in list_of_words]

def generate_images(n, images_dir):
    for i in range(n):
        xy = [0,0]
        n = random.randint(3, 10)
        new_str = list_of_words[random.randint(0, len(list_of_words)-1)]
        font = fonts[random.randint(0, len(fonts)-1)]
        font = ImageFont.truetype(font_folder_path+'/'+font, random.randint(25,40), encoding="unic")
        
        _, _, w, h = font.getbbox(new_str)
        offset = (random.randint(10, 20), random.randint(40,60))
        size = (w + offset[0], h + offset[1])
        img = Image.new("L", size=size, color= "white")

        draw = ImageDraw.Draw(img)
        xy = (offset[0]/2, offset[1]/2)
        draw.text(xy, new_str, font = font, fill = "black")

        img.save(os.path.join(images_dir, f"{i}_{new_str}.jpg"))


generate_images(50000, img_dir_train)
generate_images(500, img_dir_test)
```
